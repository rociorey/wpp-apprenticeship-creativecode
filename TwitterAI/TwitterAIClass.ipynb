```python
tweets_wellbeing = client.search_recent_tweets(
    query = "WellbeingWednesday -is:retweet",  # searches for #WPP while ignoring retweets
    max_results = 100
)

tweets_wellbeing.data
```

```python
tweets_breezy = client.search_recent_tweets(
    query = "Breezy -is:retweet",  # searches for #WPP while ignoring retweets
    max_results = 100
)

tweets_breezy.data
```

```python
breezy_tweets = []
for x in tweets_breezy.data:
    breezy_tweets.append(x.text)
    
breezy_tweets
```

```python
import random
def get_features(text):
    # create array of words
    text = text.lower()
    word_lists = text.split()
    s_words =stopwords.words('english')
    s_words = stopwords.words('spanish')
    
    #One way of doing it from the github repo
    #keywords = dict([(w, True) for w in wordsList if w not in s_words and not w.isdigit() and '/' not in w])
    
    #Another way of doing it, like we previously did in class
    characters_i_hate = ['@', '//', 'https']
    keywords = [x for x in word_lists if x not in s_words and not any(c in x for c in characters_i_hate)]
    random.shuffle(keywords)
    while len(keywords) < 3:
        keywords.append('None')
        
    
    return {'first_keyword': keywords[0], 'second_keyword': keywords[1], 'third_keyword': keywords[2]}

#get_features(wellbeing_tweets[10])

wellbeing_tweets
```

```python
# create a list of tuples containing features and the label 

features_and_labels = []

for x in wellbeing_tweets:
    features_and_labels.append((get_features(x), 'Wellbeing'))
    
for x in breezy_tweets:
    features_and_labels.append((get_features(x), 'Breezy'))
```

```python
features_and_labels
```

```python
#Always shuffle your features and labels

random.shuffle(features_and_labels)
```

```python
import math

split_num = math.floor(len(features_and_labels)* .8)
split_num
```

```python
training_set = features_and_labels[:split_num]
testing_set = features_and_labels[split_num:]
```

```python
#create our classifier
#my baseline is 50% because i hve two lbels
my_classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
my_classifier.classify(get_features('A bad tweet about how cold it has been lately'))
```

```python
# Calculate accuracy
nltk.classify.accuracy(my_classifier,testing_set)
```
