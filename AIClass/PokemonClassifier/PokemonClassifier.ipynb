```python
import pandas as pd
```

```python
path = 'pokemon_data.csv'
# Convverter is going to take a key value pair. Converters alow pandas to know that abilities is a list, not a string
pokemon = pd.read_csv(path, converters={'abilities': pd.eval})
pokemon
```

```python

#You can use the dot notation because we are using the pandas library and no space in the column
abilities = pokemon.abilities
```

```python
#Defining out features
#Key-value paira
#The keys will be the same for everyone. the values will differ
#{'first move': Overgrow, 'second-move': Chlorophyll}
```

```python
# Our first funciton

def create_features(pk_moves):
    if len(pk_moves) == 1:
        pk_moves.append('None')
    return {'first move': pk_moves[0], 'second-move': pk_moves[1]}
```

```python
# Zip is an innate python function that combines our two columns
# each element of a zip are called tuples
# syntax for tuple (thing, thing2)

features_and_labels = zip(pokemon.abilities, pokemon.type1)
```

```python
feature_sets = [(create_features(x),y) for x,y in features_and_labels]
```

```python
feature_sets
```

```python
import random
random.shuffle(feature_sets)
```

```python
len(feature_sets)
```

```python
# Figuring out how much the 80% of the lenght of feature_sets by dividing beteen .8
801* 0.8
```

```python
#Splitting our training set and testing set into a 80/20 ratio
training_set = feature_sets[:640]
testing_set = feature_sets[640:]
```

```python
import nltk

#build our first classifier
classifier = nltk.NaiveBayesClassifier.train(training_set)
```

```python
classifier.classify(create_features(['Blaze', 'Solar Power']))
```

```python
# how accurate is our classifier?
print(nltk.classify.accuracy(classifier, testing_set))
```
